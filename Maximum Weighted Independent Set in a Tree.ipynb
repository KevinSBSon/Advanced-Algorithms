```python
import sys
import threading

# To avoid stack overflow issues
sys.setrecursionlimit(10**6) 
threading.stack_size(2**26)  

class Vertex:
    def __init__(self, weight):
        self.weight = weight
        self.children = []


def ReadTree():
    global size
    print("Input: ")
    size = int(input())
    tree = [Vertex(w) for w in map(int, input().split())]
    for i in range(1, size):
        a, b = list(map(int, input().split()))
        tree[a - 1].children.append(b - 1)
        tree[b - 1].children.append(a - 1)
    return tree

D = []
parent = []

def FunParty(v, tree):
    global D, parent
    m0 = 0
    m1 = 0
    m1Fun = 0
    if D[v] == float('inf'):
        if (len(tree[v].children) == 1) and (v != 0):
            D[v] = tree[v].weight
        else:
            m1 = tree[v].weight
            for x in tree[v].children:
                if parent[x] ==-1:
                    parent[x] = v
                if parent[x] == v:
                    for y in tree[x].children:
                        if parent[y] ==-1:
                            parent[y] = x
                        if parent[y] == x:
                            m1Fun = FunParty(y, tree)
                            m1 = m1 + m1Fun
            for x in tree[v].children:
                if parent[x] == v:
                    m0 = m0 + FunParty(x, tree)
            D[v] = max(m0, m1)
    return D[v]

def MaxWeightIndependentTreeSubset(tree):
    global size, D, parent
    size = len(tree)
    if size == 0:
        return 0

    for i in range(len(D)):
        if D[i] == float('inf'):
            FunParty(i, tree)
    result = max(D)
    return result
```
